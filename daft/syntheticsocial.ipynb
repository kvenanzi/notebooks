```python
import random
from datetime import datetime, timedelta
import daft

def generate_social_media_data_fast(num_records: int) -> daft.DataFrame:
    """
    Generate social media posts data using Daft - optimized for speed
    """
    print(f"# Generating {num_records:,} social media posts...")

    base_time = datetime.now()

    post_types = ['text', 'image', 'video', 'link', 'poll']
    categories = ['technology', 'lifestyle', 'news', 'entertainment', 'sports', 'business']
    languages = ['en', 'es', 'fr', 'de', 'it', 'pt', 'ru', 'zh', 'ja', 'ko']
    devices = ['mobile', 'desktop', 'tablet']
    platforms = ['twitter', 'facebook', 'instagram', 'linkedin', 'tiktok']

    data = {
        'post_id': [f"post_{i:012d}" for i in range(num_records)],
        'user_id': [f"user_{random.randint(1000, 999999):06d}" for _ in range(num_records)],
        'username': [f"user_{random.randint(1000, 999999):06d}" for _ in range(num_records)],
        'post_type': [random.choice(post_types) for _ in range(num_records)],
        'content': [f"Sample social media post content {i} with hashtags #social #media #data"
                    for i in range(num_records)],
        'content_category': [random.choice(categories) for _ in range(num_records)],
        'created_at': [(base_time - timedelta(days=random.randint(0, 365),
                                              hours=random.randint(0, 23),
                                              minutes=random.randint(0, 59)))
                       .strftime("%Y-%m-%d %H:%M:%S") for _ in range(num_records)],
        'updated_at': [(base_time - timedelta(days=random.randint(0, 30),
                                              hours=random.randint(0, 23),
                                              minutes=random.randint(0, 59)))
                       .strftime("%Y-%m-%d %H:%M:%S") for _ in range(num_records)],
        'likes_count': [random.randint(0, 10000) for _ in range(num_records)],
        'comments_count': [random.randint(0, 1000) for _ in range(num_records)],
        'shares_count': [random.randint(0, 500) for _ in range(num_records)],
        'views_count': [random.randint(0, 1_000_000) for _ in range(num_records)],
        'is_public': [random.choice([True, False]) for _ in range(num_records)],
        'is_verified': [random.choice([True, False]) for _ in range(num_records)],
        'location': [random.choice([None, f"City_{random.randint(1, 100)}"])
                     for _ in range(num_records)],
        'language': [random.choice(languages) for _ in range(num_records)],
        'device_type': [random.choice(devices) for _ in range(num_records)],
        'platform': [random.choice(platforms) for _ in range(num_records)],
        'engagement_rate': [round(random.uniform(0.01, 0.15), 4) for _ in range(num_records)],
        'sentiment_score': [round(random.uniform(-1.0, 1.0), 3) for _ in range(num_records)],
        'spam_score': [round(random.uniform(0, 1), 3) for _ in range(num_records)],
        'has_media': [random.choice([True, False]) for _ in range(num_records)],
        'media_type': [random.choice(['None', 'image', 'video'])
                       if random.choice([True, False]) else None
                       for _ in range(num_records)],
        'media_url': [f"https://example.com/media/{random.randint(1000, 9999)}.jpg"
                      if random.choice([True, False]) else None
                      for _ in range(num_records)],
        'parent_post_id': [f"parent_{random.randint(1000, 9999):06d}"
                           if random.choice([True, False]) else None
                           for _ in range(num_records)],
        'is_repost': [random.choice([True, False]) for _ in range(num_records)],
        'original_author_id': [f"user_{random.randint(1000, 9999):06d}"
                               if random.choice([True, False]) else None
                               for _ in range(num_records)],
        'tags': [[f"tag{random.randint(1, 10)}" for _ in range(random.randint(1, 10))]
                 for _ in range(num_records)],
        'mentions': [[f"user_{random.randint(1000, 9999)}"
                      if random.choice([True, False]) else None]
                     for _ in range(num_records)],
        'urls': [f"https://example.com/link/{random.randint(1000, 9999)}"
                 if random.choice([True, False]) else None
                 for _ in range(num_records)],
        'character_count': [random.randint(10, 280) for _ in range(num_records)],
        'word_count': [random.randint(2, 50) for _ in range(num_records)],
        'hashtag_count': [random.randint(0, 10) for _ in range(num_records)],
        'is_trending': [random.choice([True, False]) for _ in range(num_records)],
        'trend_score': [round(random.uniform(0.0, 100.0), 2) for _ in range(num_records)],
    }

    df = daft.from_pydict(data)
    return df
```

```python
generate_social_media_data_fast(num_records=1000)
```

```python
df = generate_social_media_data_fast(1000)

df.write_parquet("data/social_posts.parquet", compression="snappy")
```

```python
df = daft.read_parquet("data/social_posts.parquet")
df.show(500)
```
